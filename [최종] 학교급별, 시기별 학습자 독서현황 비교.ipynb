#주제1. 2013년과 2021년의 학교급별 학생들의 독서대화 빈도 비교

```python
import csv
f=open('독서대화_학생.csv','rt', encoding='UTF8')
data=csv.reader(f, delimiter=',')

age11_2013=[]
age11_2021=[]
age12_2013=[]
age12_2021=[]
age13_2013=[]
age13_2021=[]
age14_2021=[]
age15_2021=[]
age16_2021=[]
age17_2021=[]
age18_2021=[]
age19_2021=[]

next(data)
for row in data:
  for i in range(3,8):
    if '초등학교 4학년' in row[1]:
      age11_2013.append(float(row[i]))
      age11_2021.append(float(row[i+6]))
    if '초등학교 5학년' in row[1]:
      age12_2013.append(float(row[i]))
      age12_2021.append(float(row[i+6]))
    if '초등학교 6학년' in row[1]:
      age13_2013.append(float(row[i]))
      age13_2021.append(float(row[i+6]))
    if '중학교 1학년' in row[1]:
      age14_2021.append(float(row[i+6]))
    if '중학교 2학년' in row[1]:
      age15_2021.append(float(row[i+6]))
    if '중학교 3학년' in row[1]:
      age16_2021.append(float(row[i+6]))
    if '고등학교 1학년' in row[1]:
       age17_2021.append(float(row[i+6]))
    if '고등학교 2학년' in row[1]:
      age18_2021.append(float(row[i+6]))
    if '고등학교 3학년' in row[1]:
      age19_2021.append(float(row[i+6]))
```

```python
#2021년 학교급별 학생들의 독서대화 빈도 비교

import matplotlib.pyplot as plt
plt.title('2021년 학교급별 학생들의 독서대화 빈도 비교')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age11_2021, color='orange',label='초등학교 4학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age12_2021, color='red',label='초등학교 5학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age13_2021, color='hotpink',label='초등학교 6학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age14_2021, color='green',label='중학교 1학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age15_2021, color='blue',label='중학교 2학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age16_2021, color='skyblue',label='중학교 3학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age17_2021, color='purple',label='고등학교 1학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age18_2021, color='grey',label='고등학교 2학년')
plt.plot(['매우자주함','자주하는편','보통','별로하지않음','전혀하지않음'], age19_2021, color='black',label='고등학교 3학년')
plt.legend()
plt.show()
```

#주제2. 2013년과 2021년 학교급별 학생들의 종이책+전자책 독서량 변화비교

```python
import csv
f=open('종이책전자책독서량_학생.csv','r', encoding='cp949')
data=csv.reader(f, delimiter=',')

next(data)
for row in data:
  print(row)
```

```python
import csv
f=open('종이책전자책독서량_학생.csv','r', encoding='cp949')
data=csv.reader(f, delimiter=',')

elementary=[]
middle=[]
high=[]

next(data)
for row in data:
  if '초등학교' in row[1]:
    elementary.append(float(row[3]))
    elementary.append(float(row[6]))
  if '중학교' in row[1]:
    middle.append(float(row[3]))
    middle.append(float(row[6]))
  if '고등학교' in row[1]:
    high.append(float(row[3]))
    high.append(float(row[6]))
    break
print(elementary)
print(middle)
print(high)

import matplotlib.pyplot as plt
plt.title('2013년과 2021년의 학교급별 독서량 비교')
plt.plot(['2013','2021'], elementary, color='blue',marker='o',linestyle='dashed',label='초등학교')
plt.plot(['2013','2021'], middle, color='hotpink', marker='x',linestyle='solid', label='중학교')
plt.plot(['2013','2021'], high, color='green', marker='s',linestyle='dotted', label='고등학교')

plt.legend()
plt.show()
```

#주제3. 학교급별 도서입수경로 비교 (2021년)

```python
import csv
f=open('도서입수경로_학생.csv','r', encoding='cp949')
data=csv.reader(f, delimiter=',')

next(data)
for row in data:
  print(row)
```

```python
import csv
f=open('도서입수경로_학생.csv','r', encoding='cp949')
data=csv.reader(f, delimiter=',')

elementary=[]
middle=[]
high=[]

next(data)
for row in data:
  if '초등학교' in row[1]:
    elementary=list(map(float,row[3:10]))
  if '중학교' in row[1]:
    middle=list(map(float,row[3:10]))
  if '고등학교' in row[1]:
    high=list(map(float,row[3:10]))
    break

print(elementary)
print(middle)
print(high)

import matplotlib.pyplot as plt
plt.title('초등학생의 도서입수경로(2021년)')
plt.axis('equal')
label=['직접 구입','집에 있는 책','학교 도서관','부모님 추천','학교밖도서관','주변사람에게 빌림','선물']
color=['darkmagenta','deeppink','hotpink','pink','lightcoral','salmon','seashell']
plt.pie(elementary, labels=label, autopct='%1.0f%%',colors=color)
plt.legend()
plt.show()

plt.title('중학생의 도서입수경로(2021년)')
plt.axis('equal')
label=['직접 구입','집에 있는 책','학교 도서관','부모님 추천','학교밖도서관','주변사람에게 빌림','선물']
color=['forestgreen','limegreen','darkgreen','seagreen','mediumseagreen','lightseagreen','green']
plt.pie(middle, labels=label, autopct='%1.0f%%',colors=color)
plt.legend()
plt.show()

plt.title('고등학생의 도서입수경로(2021년)')
plt.axis('equal')
label=['직접 구입','집에 있는 책','학교 도서관','부모님 추천','학교밖도서관','주변사람에게 빌림','선물']
color=['cornflowerblue','royalblue','darkblue','slateblue','darkslateblue','midnightblue','lavender']
plt.pie(high, labels=label, autopct='%1.0f%%',colors=color)
plt.legend()
plt.show()
```
